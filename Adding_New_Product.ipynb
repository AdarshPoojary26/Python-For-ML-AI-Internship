# Add New Item Into Inventory

```python
product_id = str(input("Enter the Product ID :"))
name = str(input("Enter the Product Name :"))
product_price = int(input("Enter the Product Price :"))
quantity = int(input("Enter the Product Quantity :"))
expire_date = str(input("Enter the product Expire Date :"))
description = str(input("Enter product Description:"))                  
                  
items[product_id] = {"Name": name, "Product price": product_price, "Quantity": quantity, "Expire_date": expire_date, "Description": description}

js = json.dumps(items)

fd = open("record.json",'w')
fd.write(js)
fd.close()
```

```python
fd = open("record.json",'r')
txt = fd.read()
fd.close()

items = json.loads(txt)
```
